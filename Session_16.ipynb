```python
print("Numerical columns:")
for i in df.columns:
  if df[i].dtypes!="object":
    print(i)

print("\nCategorical columns:")
for i in df.columns:
  if df[i].dtypes=="object":
    print(i)
```

```python
plt.subplot(2,2,1)
sns.histplot(df["age"])
plt.title("Age")
plt.xlabel("Age")
plt.ylabel("count")

plt.subplot(2,2,2)
sns.histplot(df["bmi"])
plt.title("BMI")
plt.xlabel("bmi")
plt.ylabel("count")

plt.subplot(2,2,3)
sns.histplot(df["children"])
plt.title("Children")
plt.xlabel("children")
plt.ylabel("count")

plt.subplot(2,2,4)
sns.histplot(df["expenses"])
plt.title("expenses")
plt.xlabel("expenses")
plt.ylabel("count")
plt.tight_layout()
plt.show()


```

```python
plt.subplot(2,2,1)
sns.countplot(x=df["sex"])
plt.title("count of Sex")
plt.xlabel("sex")
plt.ylabel("count")

plt.subplot(2,2,2)
sns.countplot(x=df["smoker"])
plt.title("count of smoker")
plt.xlabel("smoker")
plt.ylabel("count")

plt.subplot(2,2,3)
sns.countplot(x=df["region"])
plt.title("count of region")
plt.xlabel("region")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```

```python
co=df.corr(numeric_only=True)
sns.heatmap(co,annot=True)
```

```python
print("Average insurance Expenses:")
print(df["expenses"].mean())
```

```python
print("Maximum Expense:")
print(df["expenses"].max())
print("\nMinimum Expense")
print(df["expenses"].min())
```

```python
print("Average charges by smokers")
print(df.groupby("smoker")["expenses"].mean())
```

```python
sns.boxplot(x="sex",y="expenses",hue="smoker",data=df)
plt.title("Distribution of Expenses by sex ans smoker")
plt.xlabel("Expenses")
plt.ylabel("Sex")
plt.show()

```
